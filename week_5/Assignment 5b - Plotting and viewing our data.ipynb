```python
%matplotlib inline
import numpy as np
import pandas as pd

from IPython.display import display, HTML

from matplotlib import pyplot as plt
from matplotlib.pyplot import figure
```

```python
df =  pd.read_csv('my_ass5_dataset.csv')
days =  len(df.index)
print(days)
display(df)
```

### 2.1 Plotting the data

```python
def plot_data():
    """This function plots the data of each column in the csv file"""
    x = np.linspace(0, days-1, days)
    plt.xlim(-10, days)
    plt.xlabel("Day")
    plt.ylabel("Population")
    for column in df.columns: 
        plt.plot(x, df[column])
        
    plt.legend(df.columns)
    plt.savefig("Dataplot1.jpg")

figure(figsize=(8, 8), dpi = 80)
plot_data()
```

### 2.3 some other interesting plot

```python
# Bar Plot
def bar_plot():
    """This function will plot the data found the sick column"""
    figure(figsize=(8, 8), dpi = 80) # A neat trick to get larger images, google it for more info
    
    df["Sick"].plot.bar(rot = 0)
    
    plt.xlabel("Day")
    plt.ylabel("Population")
    plt.legend(["Amount of people sick each day"])
    plt.title("Barchart of sick Population")
    
    plt.savefig("Barplot1.jpg")
    
bar_plot()
```

```python
# Multiple Plots
def mult_plots():
    nrows=1
    ncols=2
    fig, axes = plt.subplots(nrows, ncols,figsize=(8, 3), dpi = 80)
    plt.subplots_adjust(left=0.1,
                    bottom=0.1,
                    right=0.9,
                    top=0.9,
                    wspace=0.4,
                    hspace=0.4)
    axes[0].margins(0.01)
    axes[1].margins(0.01)
    df["Sick"].plot(ax=axes[0], color="red", title="People sick",xlabel="Day",ylabel="Population")
    df["Recoverd"].plot(ax=axes[1], color="green", title="People recoverd",xlabel="Day",ylabel="Population")
    plt.savefig("Subplot1.jpg")

mult_plots()
```

```python
# Creating the scatter plot
x = df["Recoverd"]
y = df["Vaccinated"]
area = df["Susceptible"]

fig, ax = plt.subplots(figsize=(8, 8), dpi=80)
ax.set_xlabel("Recoverd", fontsize=15)
ax.set_ylabel("Vaccinated", fontsize=15)
ax.set_title("Area represents Susceptible")
scatter = ax.scatter(x, y, s=area, c=area, cmap='coolwarm', alpha=0.5)

# Adding a color bar
cbar = plt.colorbar(scatter, ax=ax, label='Susceptible')

plt.savefig("scatter_plot.png")
plt.show()

```
