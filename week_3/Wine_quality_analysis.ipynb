```python
def count_nan(df):
    # For each column in the dataset perform the isna() operation and summarize the missing samples
    for col in df.columns:
        # Calculate the number of missing values (isna().sum()) and the total percentage
        missing_count = df[col].isna().sum()
        missing_percentage = (missing_count / len(df)) * 100
        # Print the result with formatted string
        print(f"The column {col} is missing {missing_percentage:.2f}% of data!")



count_nan(df)
```

### Transforming Data

Normalization transforms the data so that all values in each column range between 0 and 1. 

This makes comparing features easier since theyâ€™re on the same scale. 

We do this by subtracting the minimum value from each data point and dividing by the range (maximum minus minimum). The idea is to make the data consistent across columns, which helps in analysis and machine learning.

```python
def normalize_values(col_value, min_val, max_val):
    # Normalize the value between 0 and 1 using min-max scaling
    normalized_val = (col_value - min_val) / (max_val - min_val)
    return normalized_val

def main():
    for column in df.columns:
        # Calculate the minimum and maximum values for each column
        min_val = df[column].min()
        max_val = df[column].max()
        # Apply the normalization function to each column
        df[column] = df[column].apply(normalize_values, args=(min_val, max_val))

main()

# Display the statistical summary of the normalized dataset
df.describe()
```

# Statistical tool : The 5-number summary

The 5-number summary is a valuable statistical tool for understanding data distribution and identifying outliers. It consists of five key values,

1. the minimum (smallest data point),
2. the first quartile 25% (Q1),
3. the median 50% (Q2 or the second quartile),
4. the third quartile 75% (Q3), and
5. the maximum (largest data point).


### To detect outliers
we focus on the interquartile range (IQR)

calculated as:
IQR = Q3 - Q1
This IQR represents the middle 50% of the data.

Outliers are identified as values that fall significantly lower fence and higher fence.

Lower fence = Q1 - 1.5 * IQR
Higher fence = Q3 + 1.5 * IQR

Essentially, the 5-number summary with IQR helps us pinpoint unusual data points that might deviate from the overall pattern and warrant closer investigation.
